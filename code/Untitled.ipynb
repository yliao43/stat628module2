```python
#create wordtoid, idtoword, document
def createDict(cut_docs):
        word2id = dict()
        wordnum = 0
        for i, doc in enumerate(cut_docs):
            for j, word in enumerate(doc):
                wordnum += 1
                if word not in word2id:
                    word2id[word] = len(word2id)
                cut_docs[i][j] = word2id[word]
        return word2id, dict(zip(word2id.values(), word2id.keys())), cut_docs, wordnum
```

```python
def reSamppling(nd, nw, ndsum, nwsum):
        pk = np.ndarray([K])
        for i in range(K):
            # gibbs采样公式
            pk[i] = float(nd + alpha)*(nw +beta)/\
                    ((ndsum + alpha*K)*(nwsum + beta*V))
            if i > 0:
                pk[i] += pk[i-1]

        # 轮盘方式随机选择主题
        u = random.random()*pk[K-1]
        for k in range(len(pk)):
            if pk[k]>=u:
                return k
```

```python
def updatePara():
        for d in range(D):
            for k in range(K):
                theta[d,k] = float(nd[d,k] + alpha)/\
                                (ndsum[d] + alpha*K)

        for k in range(K):
            for v in range(V):
                phi[k,v] = float(nw[k,v] + beta)/\
                                (nwsum[k] + beta*K)
```

```python
#gibbs sampling
for iter in range(max_iter_num):
    for i, doc in enumerate(cutdocs):
        for j, word_id in enumerate(doc):
            theme = z[i,j]
            nd_t = nd[i,theme] - 1
            nw_t = nw[theme,word_id] - 1
            ndsum_t = ndsum[i] - 1
            nwsum_t = nwsum[theme] - 1
            # 重新给词选择新的主题
            new_theme = reSamppling(nd_t, nw_t, ndsum_t, nwsum_t)
            
            nd[i,theme] -= 1
            nw[theme,word_id] -= 1
            nwsum[theme] -= 1
                    
            nd[i,new_theme] += 1
            nw[new_theme,word_id] += 1
            nwsum[new_theme] += 1
            z[i,j] = new_theme
    sys.stdout.write('\rIteration:{0} done!'.format(iter+1))
    sys.stdout.flush()

```
